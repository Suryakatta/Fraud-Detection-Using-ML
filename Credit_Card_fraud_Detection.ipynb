```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
dataset=pd.read_csv('/content/credit_data.csv')
```

```python
 dataset.head()
```

```python
dataset.shape
```

```python
dataset['Class'].value_counts()
```

```python
dataset.dropna(subset=['Class'], inplace=True)
dataset['Class']=dataset['Class'].astype(int)
```

```python
dataset.head()
```

```python
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
legit.head()
```

```python
legit.shape
```

```python
legit_sample=legit.sample(n=250)
```

```python
legit_sample.head()
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.shape
```

```python
new_dataset.isnull().sum()
```

```python
new_dataset['Class'].value_counts()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
```

```python
Y=new_dataset['Class']
```

```python
X.head()
```

```python
Y.head()
```

```python
scaler=StandardScaler()
```

```python
Standard_data=scaler.fit_transform(X)
```

```python
print(Standard_data)
```

```python
X=Standard_data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction=model.predict(X_train)
accuracy_for_training=accuracy_score(Y_train,training_data_prediction)
```

```python
print(accuracy_for_training)
```

```python
testing_data_prediction=model.predict(X_test)
accuracy_for_testing=accuracy_score(testing_data_prediction,Y_test)
```

```python
print(accuracy_for_testing)
```
